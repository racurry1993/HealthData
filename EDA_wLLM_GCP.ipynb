```python
string_columns = [
    'Date',
    'locationName',
    'trainingEffectLabel',
    'aerobicTrainingEffectMessage',
    'anaerobicTrainingEffectMessage',
    'typeKey_clean',
    'startTimeCat',
    'ageGroup',
    'sleepScoreInsight',
    'sleepScoreFeedback',
    'stressQualifier',
    'ActivityStartHour',
    'activityType',
    'day_of_week'
]

# Iterate through the list and safely convert each column to string
for col in string_columns:
    if col in df.columns:
        df[col] = df[col].astype(str)

# Example from a previous step: If a numeric column has a mixed type, you can use pd.to_numeric
if 'value' in df.columns:
    df['value'] = pd.to_numeric(df['value'], errors='coerce')
    df['value'] = df['value'].fillna(0)

def df_togcp(df, svc_acct_key, table_id, write_disp, schema=None):
    
    # Write Disposition:
    #    WRITE_TRUNCATE = If the table already exists, BigQuery overwrites the data, removes the constraints, and uses the schema from the query result.
    #    WRITE_APPEND = If the table already exists, BigQuery appends the data to the table
    #    WRITE_EMTPY = If the table already exists and contains data, a 'duplicate' error is returned in the job result.
    # Docs: https://cloud.google.com/bigquery/docs/reference/rest/v2/Job
    
    # Schema [Optional]:
    #   Pass a list of columns whose data type you will explicitly define
    #   See example here: https://cloud.google.com/bigquery/docs/samples/bigquery-load-table-dataframe
    
    from google.cloud import bigquery
    os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = svc_acct_key
    client = bigquery.Client()
    
    
    job_config = bigquery.LoadJobConfig(write_disposition = write_disp)
    
    if schema is not None:
        job_config = bigquery.LoadJobConfig(schema=schema, write_disposition=write_disp)
    
    job = client.load_table_from_dataframe(df, table_id, job_config=job_config)
    job.result()   # This line waits for load to complete
    
    table = client.get_table(table_id)  
    print(
        "Loaded {} rows and {} columns to {}".format(
            table.num_rows, len(table.schema), table_id
        )
    )


df_togcp(df, key, 'garminuserdata.garminuserdata.garmin_activity_data', write_disp='WRITE_TRUNCATE')
```

```python
def gcp_todf(query, svc_acct_key):
    
    from google.cloud import bigquery
    os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = svc_acct_key
    client = bigquery.Client()
    
    query_job = client.query(query)
    df = query_job.to_dataframe()
    
    return df
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt


# Select the key columns for our analysis
# 'restingHeartRate' is our dependent variable (what we want to predict).
# 'sleepingSeconds' and 'totalSteps' are our key independent variables.
# We'll use 'activeKilocalories' as a proxy for "working out."
# We'll also use 'averageStressLevel' as another potential factor.
analysis_cols = ['restingHeartRate', 'remSleepHours', 'awakeSleepPercentage', 'sleepingSeconds', 'totalSteps',
                  'ActivityPerformedToday', 'activeKilocalories', 'averageStressLevel', 'day_of_week','highStressPercentage',
                  'mediumStressPercentage','lowStressPercentage','uncategorizedStressPercentage','activityStressPercentage',
                  'restStressPercentage','highStressDuration','floorsAscended','floorsDescended','lastSevenDaysAvgRestingHeartRate',
                  'charged','drained']
df_eda = df[analysis_cols].copy()

# Drop rows with any missing values to ensure the regression model runs smoothly
df_eda.dropna(inplace=True)

# --- FIX START ---
# Force conversion to float. This is the fix for the TypeError.
# It ensures all values are numeric, preventing the ufunc error.
for col in analysis_cols:
    df_eda[col] = pd.to_numeric(df_eda[col], errors='coerce')

# Drop any rows that may have become NaN after coercion
df_eda.dropna(inplace=True)
# --- FIX END ---

# Convert sleep duration from seconds to hours for easier interpretation
df_eda['sleep_hours'] = df_eda['sleepingSeconds'] / 3600

# Create a rolling 7-day average for key metrics
# This directly addresses your question about a "7-day" impact.
df_eda['sleep_hours_7d_avg'] = df_eda['sleep_hours'].rolling(window=7).mean().shift(1)
df_eda['totalSteps_7d_avg'] = df_eda['totalSteps'].rolling(window=7).mean().shift(1)
df_eda['activeKilocalories_7d_avg'] = df_eda['activeKilocalories'].rolling(window=7).mean().shift(1)

# Drop any rows that now have NaNs due to the rolling average calculation
df_eda.dropna(inplace=True)

# --- NEW FIX START ---
# Explicitly convert the columns used in plotting and regression to float
# to prevent any potential type issues.
df_eda['restingHeartRate'] = df_eda['restingHeartRate'].astype(float)
df_eda['sleep_hours_7d_avg'] = df_eda['sleep_hours_7d_avg'].astype(float)
df_eda['totalSteps_7d_avg'] = df_eda['totalSteps_7d_avg'].astype(float)
df_eda['activeKilocalories_7d_avg'] = df_eda['activeKilocalories_7d_avg'].astype(float)

# --- NEW FIX END ---

print(f"\nDataFrame prepared for analysis. Shape: {df_eda.shape}")
print("First 5 rows of the prepared data:")
print(df_eda.head())

# -----------------------------------------------------------------------------
# Step 2: Exploratory Data Analysis with Visuals
# -----------------------------------------------------------------------------

# Set up the plotting style
sns.set_style("whitegrid")
plt.style.use("ggplot")

# Plotting the Resting Heart Rate over time to see trends
plt.figure(figsize=(14, 6))
plt.plot(df_eda.index, df_eda['restingHeartRate'], color='red', linewidth=2, label='Resting Heart Rate')
plt.title('Resting Heart Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Heart Rate (bpm)')
plt.legend()
plt.tight_layout()
plt.show()

# Create scatter plots to visualize relationships
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)
fig.suptitle('Impact of Key Metrics on Resting Heart Rate', fontsize=18)

# Scatter plot for Sleep vs. Resting HR
sns.regplot(ax=axes[0], x='sleep_hours_7d_avg', y='restingHeartRate', data=df_eda,
            scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
axes[0].set_title('7-Day Avg Sleep vs. Resting HR')
axes[0].set_xlabel('Average Sleep (Hours)')
axes[0].set_ylabel('Resting Heart Rate (bpm)')

# Scatter plot for Steps vs. Resting HR
sns.regplot(ax=axes[1], x='totalSteps_7d_avg', y='restingHeartRate', data=df_eda,
            scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
axes[1].set_title('7-Day Avg Steps vs. Resting HR')
axes[1].set_xlabel('Average Steps')
axes[1].set_ylabel('') # Y-axis label is shared

# Scatter plot for Active Calories vs. Resting HR
sns.regplot(ax=axes[2], x='activeKilocalories_7d_avg', y='restingHeartRate', data=df_eda,
            scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
axes[2].set_title('7-Day Avg Active Calories vs. Resting HR')
axes[2].set_xlabel('Average Active Calories')
axes[2].set_ylabel('') # Y-axis label is shared

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Create a heatmap to visualize correlations
correlation_matrix = df_eda[['restingHeartRate', 'sleep_hours_7d_avg', 'totalSteps_7d_avg', 'activeKilocalories_7d_avg', 'averageStressLevel']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Key Health Metrics')
plt.show()

# -----------------------------------------------------------------------------
# Step 3: Linear Regression Analysis
# -----------------------------------------------------------------------------

# Define the dependent and independent variables
Y = df_eda['restingHeartRate']
X = df_eda[['sleep_hours_7d_avg', 'totalSteps_7d_avg', 'activeKilocalories_7d_avg', 'averageStressLevel']]
X['averageStressLevel'] = X['averageStressLevel'].astype('float64')
# Add a constant to the independent variables for the regression model
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(Y, X).fit()

# Print the regression summary
print("\n" + "="*70)
print("             Linear Regression Model Summary")
print("="*70)
print(model.summary())
print("="*70)

# -----------------------------------------------------------------------------
# Step 4: Interpreting the Regression Results
# -----------------------------------------------------------------------------

# The coefficients tell us the change in resting HR for a one-unit change in each variable.
# For example, a coefficient of -1.5 for 'sleep_hours_7d_avg' would mean that for
# every 1-hour increase in average sleep (over 7 days), the resting HR is predicted to
# decrease by 1.5 bpm, holding all other variables constant.

# You can find these values in the 'coef' column of the summary table.
# A negative coefficient for a variable like `sleep_hours_7d_avg` indicates an
# inverse relationship: as sleep increases, resting HR tends to decrease.
# The p-value (P>|t|) for each coefficient indicates if the relationship is
# statistically significant. A p-value less than 0.05 is generally considered significant.

print("\n")
print("Interpretation of Coefficients:")
print("The 'coef' column in the summary table shows the estimated change in restingHeartRate for each unit increase in a predictor variable.")
print("- A negative coefficient for 'sleep_hours_7d_avg' suggests that more sleep is associated with a lower resting heart rate.")
print("- A positive coefficient for 'totalSteps_7d_avg' might suggest that more activity is associated with a slightly higher resting heart rate, or it could be related to the timing of the measurement.")
print("- The R-squared value indicates how much of the variance in resting heart rate is explained by the model's variables. A higher R-squared (closer to 1) means the model is a better fit for the data.")
```
